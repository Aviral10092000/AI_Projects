```python
import openai, os
```

```python
OPENAI_API_KEY = os.environ["OPENAI_API_KEY"]
```

```python
def get_response(prompt, temperature=0, model="gpt-4o-mini"):
    
    messages = [{"role":"user","content":prompt}]
    response = openai.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature)
    
    return response.choices[0].message.content
```

```python
first_ticket= '''I ordered a book two weeks ago, and it still hasn't arrived. The tracking information is not updating either.
'''
```

```python
user_prompt = f'''classify the customer support tickets into broad categories such as
"Delivery Issues", "Product Quality", and "General Feedback"
{first_ticket}'''

print(get_response(user_prompt))
```

```python
fourth_ticket = '''I purchased a laptop from your store, and it stopped working within a week. I tried reaching out to your support team, but nobody responded. Please help!'''
```

```python
user_prompt=f'''
Generate a response for a customer support ticket related to a user experiencing in Casual and empathetic tone

Customer Support Ticket:{fourth_ticket}
'''
causal_empathetic = get_response(user_prompt)
print(causal_empathetic)
```
