```python
import autogen

config_list_gpt4o = autogen.config_list_from_json(
    "OAI_CONFIG_LIST.txt"
)
```

```python
gpt4o_config = {
    "seed": 42,  # change the cache_seed for different trials
    "temperature": 0,
    "config_list": config_list_gpt4o,
    "timeout": 120,
}

user_proxy = autogen.UserProxyAgent(
    name="Admin",
    system_message="A human admin. Interact with a Business_Owners_and_Managers to discuss the plan of execution. This plan needs to be approved by this admin (Boss).",
    code_execution_config=False
)

planner = autogen.AssistantAgent(
    name='Business_Owners_and_Managers', 
    system_message='Clothing store owners or managers in areas with high competition can use this report to identify key competitors, analyze peak customer hours, and adjust their own strategies to improve market position.',
    llm_config = gpt4o_config
)

engineer = autogen.AssistantAgent(
    name="Marketing_and_Strategy_Teams",
    llm_config=gpt4o_config,
    system_message="""Marketing teams responsible for local engagement can leverage insights on footfall trends to optimize promotions during high-traffic periods. Additionally, competitive data can help strategists shape targeted campaigns, aligning with or counteracting competitor timings."""
)
scientist = autogen.AssistantAgent(
    name="RealEstate_and_Location_Analysts",
    llm_config=gpt4o_config,
    system_message="""Location-based data on competitor footfall and busy hours can also assist analysts in recommending ideal store locations for future openings, as well as identifying high-value times for new store launches.""",
)
critic = autogen.AssistantAgent(
    name="Investors_and_Market_Analysts",
    system_message="Investors considering opportunities in retail can use competitor analysis to assess market saturation, business potential, and potential risks in high-traffic areas like Koramangala, Bangalore.",
    llm_config=gpt4o_config,
)
```

```python
groupchat = autogen.GroupChat(
    agents=[user_proxy, planner, engineer, scientist, critic],
    messages=[],
    max_round=50
)

manager = autogen.GroupChatManager(groupchat=groupchat, llm_config=gpt4o_config)
```

```python
output_report = user_proxy.initiate_chat(manager,
                                         system_message=". The system integrates a language model with a search tool to provide real-time data on competitors, including store footfall and busiest times, giving actionable insights to enhance business strategies.")
```
